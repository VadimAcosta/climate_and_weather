## **The Daily Weather in NYC** 

```python
import pandas as pd
```

Importing the dataset csv

```python
df = pd.read_csv("data/NOAA_NY_Dailies.csv")
```

Taking a look around

```python
df.info()
```

## **Removing Wind Data**

```python
df.columns
```

```python
df = df.drop(columns = ['WDF2', 'WDF5', 'WESD', 'WESF', 'WSF2', 'WSF5', 'WT01', 'WT02',
       'WT03', 'WT04', 'WT05', 'WT08', 'WT11'])
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

## Renaming the Columns

```python
df.columns
```

```python
df.rename(columns={'TOBS': 'TOBS_temp_at_time', 'AWND': 'AWND_avg_daily_wind'}, inplace=True)
```

```python
df.info()
```

## Visualize

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,3))
plt.plot(df['DATE'],df['TMAX'])
```
